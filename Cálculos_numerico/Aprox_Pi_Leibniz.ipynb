```python
import matplotlib.pyplot as plt

resultados = []
num_iteracoes = range(60, 500, 10)

for lim in num_iteracoes:
    soma = 0.0
    for i in range(lim):
        soma += ((-1)**i) / (2 * i + 1)
    pi_aprox = 4 * soma
    resultados.append(pi_aprox)
    print(f"Iteração {lim}: {pi_aprox:.14f}")

plt.plot(num_iteracoes, resultados)
plt.xlabel('Número de Iterações')
plt.ylabel("Aproximação de π")
plt.axhline(y=3.14159265358979323846, color='r',
            linestyle='--', label='π verdadeiro')
plt.legend()
plt.show()
```
