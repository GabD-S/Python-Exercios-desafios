```python
import numpy as np
import matplotlib.pyplot as plt

s_n = [0.5, 0.5]
ini = 0
n_steps = 1000

positions = [ini]

trajectory = []

nm = np.random.random(n_steps)
ental = nm < s_n[0]
leva = nm > s_n[1]

for alto, ental in zip(ental, leva):
    down = alto and positions[-1] > -4
    up = ental and positions[-1] < 4
    positions.append(positions[-1] - down + up)
    if len(positions) % 10 == 0:
        trajectory.append(positions[-1])

trajectory = np.array(trajectory)

mean_value = np.mean(trajectory)
std_deviation = np.std(trajectory)

print("Valor Médio:", mean_value)
print("Desvio Padrão:", std_deviation)

plt.plot(trajectory)
plt.axhline(mean_value,std_deviation, color='red', linestyle='--')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import cmath

mati_demonio = np.array([
    [3, 1 + 1j, 1, 0],
    [1 - 1j, 4, 0, 2],
    [1, 0, 4, -1j],
    [0, 2, 1j, 1]
], dtype=complex)

w,psi = np.linalg.eig(mati_demonio)
autova = [complex(w[i].real,w[i].imag) for i in range(len(w))]

autova = []
for i in range(len(psi)):
    autova.append(psi[:,i].reshape((4,1)))

for i, val in enumerate(autova):
    print(f"auto_valo{i+1}--> {val}")
    print()

for i,vec in enumerate(autova):
    print(f"\nPSI_{i+1}vec")
    print(vec)
```

```python
import numpy as np

H = np.array([[3, 1 + 1j, 1, 0],
              [1 - 1j, 4, 0, 2],
              [1, 0, 4, -1j],
              [0, 2, 1j, 1]])

autovalores, autovetores = np.linalg.eig(H)
for i in range(len(autovalores)):
    print(f"Autovalor {i+1}: {autovalores[i]}")
    print(f"Autovetor {i+1}: {autovetores[:, i]}")
    print()

for i in range(len(autovalores)):
    lado_esquerdo = autovalores[i] * autovetores[:, i]
    lado_direito = np.dot(H, autovetores[:, i])
    diferenca = lado_esquerdo - lado_direito
    if np.all(np.abs(diferenca) < 1e-10):
        print("correto")
    else:
        print("errado")
```
